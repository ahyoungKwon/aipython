```python
# 실습문제 1-1 데이터프레임 생성
df = pd.DataFrame({"요일": ['월', '화', '수', '목', '금', '토', '일'],
                   '공부시간': [2, 3, 1, 5, 4, 2, 1],
                   '성적': [70, 75, 65, 85, 80, 70, 60]})
df
```

```python
# 1-2 요일별 공부 시간을 선그래프로 나타내기 (x축 = 요일, y축 = 공부시간)
# show 안해도 보이네
plt.plot(df['요일'], df['공부시간'])
```

```python
# 1-3 요일별 성적을 선 그래프로 나타내기 (x축 = 요일. y축 = 성적)
plt.plot(df['요일'], df['성적'])
```

```python
# 실습 2 - 산점도 그래프 그리기 - x축 = 공부시간, y축 = 성적

plt.scatter(df['공부시간'], df['성적'])
```

```python
# 실습 3 - numpy를 이용하여 생성한 값에 대한 산점도 그래프 그리기
import numpy as np

x = np.arange(1, 11)
y = np.random.randint(1, 30, 10)

plt.scatter(x, y)
```

```python
# 실습 4-1 도시별 인구 수를 수직 막대그래프로 시각화하기
import pandas as pd

data = {'도시': ['서울', '부산', '인천', '대구'], '인구 수': [9904312, 3448737, 2890451, 2466052]}
df = pd.DataFrame(data)

plt.bar(df['도시'], df['인구 수'])
```

```python
# 실습 4-2 그레프의 제목을 도시별 인구 수로 설정
plt.bar(df['도시'], df['인구 수'])
plt.title('도시별 인구 수')
```

```python
# 4-3 x축 라벨을 '도시'로, y축 라벨을 '인구 수'로 설정
plt.bar(df['도시'], df['인구 수'])
plt.title('도시별 인구 수')
plt.xlabel('도시')
plt.ylabel('인구 수')
```
