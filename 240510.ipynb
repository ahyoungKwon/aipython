```python
# 실습 1-1 데이터프레임 형성

df = pd.DataFrame({'학생': ['John', 'Alice', 'Bob', 'Emily'],
                   '수학': [90, 87, 78, 92],
                   '영어': [85, 91, 80, 88],
                   '과학': [88, 89, 85, 82]})
```

```python
# 1-2 요약 통계 정보 출력

print(df.describe())
```

```python
# 1-3 수학 점수가 가장 높은 학생의 점수 출력

print(df[df['수학'] == df['수학'].max()])
```

```python
# 1-4 수학 점수가 90점 이상인 학생의 정보 출력

print(df[df['수학']>=90])
```

```python
# 실습 2-1 박스오피스 파일 데이터프레임 생성

boxoffice = pd.read_excel('BoxOffice.xlsx')
boxoffice
```

```python
# 2-2 대표국적별 영화 편수 집계

a = boxoffice['대표국적'].value_counts()
print(a)
```

```python
# 2-3 대표국적이 영국인 데이터 출력
print(boxoffice[boxoffice['대표국적']=='영국'])
```

```python
# 2-4 대표국적별 최상위 순위 출력 - 힌트:min

print(boxoffice.groupby('대표국적')['순위'].min())
```
